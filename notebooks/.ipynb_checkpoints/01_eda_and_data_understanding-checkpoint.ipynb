```python
sns.countplot(x='Churn', data=df)
plt.title("Churn Distribution")
plt.show()
```

```python
numerical_cols = [
    'Age',
    'Annual_Income_USD',
    'Spending_Score',
    'Total_Purchases',
    'Avg_Purchase_Value',
    'Satisfaction_Score',
    'Website_Visits_Last_Month',
    'Avg_Time_Per_Visit_Minutes',
    'Support_Tickets_Last_6_Months',
    'Referred_Friends'
]

numerical_cols
```

```python
df.groupby('Churn')[numerical_cols].mean()
```

### Key Observations
- Churned customers exhibit slightly lower average purchase values and total purchases, indicating weaker revenue contribution.
- Engagement metrics show minimal standalone differences, suggesting the need for combined engagement features.
- Churned customers tend to raise more support tickets, pointing to potential dissatisfaction or friction.
- Referral activity is marginally lower among churned customers, indicating weaker customer advocacy.


```python
sns.boxplot(x='Churn', y='Total_Purchases', data=df)
plt.title("Total Purchases vs Churn")
plt.show()
```

### Total Purchases vs Churn
- The median total purchases are similar for churned and non-churned customers.
- The distribution of purchases shows significant overlap between the two groups.
- This suggests that total purchase count alone is not a strong churn indicator and should be combined with other engagement and satisfaction features.


```python
sns.boxplot(x='Churn', y='Satisfaction_Score', data=df)
plt.title("Satisfaction Score vs Churn")
plt.show()
```

### Satisfaction Score vs Churn
- The median satisfaction scores for churned and non-churned customers are very similar.
- There is substantial overlap in the distribution of satisfaction scores across both groups.
- This indicates that satisfaction score alone is not sufficient to explain churn and should be analyzed in combination with other features such as support tickets and engagement metrics.


```python
sns.boxplot(x='Churn', y='Support_Tickets_Last_6_Months', data=df)
plt.title("Support Tickets vs Churn")
plt.show()
```

### Support Tickets vs Churn
- Churned customers tend to raise slightly more support tickets compared to retained customers.
- High numbers of support tickets are more frequent among churned customers.
- This suggests that unresolved service issues may contribute to customer churn.


```python
categorical_cols = [
    'Gender',
    'Membership_Status',
    'Preferred_Payment_Method',
    'Region'
]

categorical_cols
```

```python
df.groupby('Membership_Status')['Churn'].mean().sort_values(ascending=False)
```

```python
df.groupby('Gender')['Churn'].mean()
```

```python
df.groupby('Preferred_Payment_Method')['Churn'].mean().sort_values(ascending=False)
```

```python
df.groupby('Region')['Churn'].mean().sort_values(ascending=False)
```

```python
sns.barplot(x='Membership_Status', y='Churn', data=df)
plt.title("Churn Rate by Membership Status")
plt.xticks(rotation=45)
plt.show()
```

### Categorical Feature Insights (Churn Rate)
- Membership status shows the clearest pattern: higher-tier members (Platinum/Gold) churn less than Silver/Bronze.
- Gender and payment method show smaller churn differences, indicating limited standalone predictive power but potential value when combined with other features.
- Region differences are minimal, suggesting churn is relatively consistent across geographies in this dataset.
